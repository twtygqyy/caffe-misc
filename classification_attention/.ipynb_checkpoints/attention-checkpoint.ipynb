```python
# -*- coding: UTF-8 ?*-
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import cv2
import caffe
```

Load the model, and some other configs

```python
caffe.set_mode_gpu()
caffe.set_device(0)

width = 224
height = 224

net = caffe.Classifier('ResNet-50-deploy.prototxt',\
                       'ResNet-50-model.caffemodel',\
                       mean=np.array([104,117,123]),\
                       channel_swap=(2,1,0),raw_scale=255,image_dims=(width, height))
```

Load the label name of each class.

```python
synset = [l.strip() for l in open('synset_words.txt').readlines()]
```

Load the image, convert into the data for caffe's input format (From BGR to RGB and normalize the image into 0-1)

```python
image = cv2.imread('ILSVRC2012_val_00043108.JPEG')
rgb = cv2.cvtColor(cv2.resize(image, (width, height)), cv2.COLOR_BGR2RGB)
image = np.array(image) 
image = image.astype(np.float32)/256.
```

Feed the image data to network

```python
prediction = net.predict([image])
```

Get the feature map of last conv layer before the average pooling layer
Get the weights for final fc layer

```python
conv_feat_map = net.blobs['res5c'].data
B, C, H, W = conv_feat_map.shape

weight_fc = net.params['fc1000'][0].data #weight without bias
```

Get the top_1 prediction

```python
top_1 = net.blobs['prob'].data[0].flatten().argsort()[-1:-2:-1]
weight_fc_top_1 = weight_fc[top_1[0],:]
```

```python
Attention_map visualization
```

```python
detection_map = weight_fc_top_1.dot(conv_feat_map.reshape(C, H*W))
detection_map = detection_map.reshape(H, W)

heat_map = cv2.resize(detection_map, (width, height))
max_response = detection_map.mean()
heat_map /= heat_map.max()

im_show = rgb.astype(np.float32)/255*0.3 + plt.cm.jet(heat_map/heat_map.max())[:, :, :3]*0.7

print 'Top 1: %s, max_response=%.4f' % (synset[top_1[0]], max_response)
plt.imshow(im_show)
plt.show()
```
